```python

columns = list(rt_weather.columns)

def count_M(columns):
    ls_M_counting = []
    ls_M_columns = []
    ls_all_counting = []
    for i in columns:
        a = rt_weather[i].value_counts()
        b = rt_weather[i].name
        if 'M' in a:
            ls_M_counting.append(a['M'])
            ls_M_columns.append(b)
            ls_all_counting.append(a.sum())
    return ls_M_counting, ls_M_columns, ls_all_counting
        
    
```

```python
k = count_M(columns)
k = np.array(k).T
k
```

```python
columns = ['M', 'column', 'All']
dfdf = pd.DataFrame(data = k, columns=columns) 
dfdf['M'] = dfdf['M'].astype(int)
dfdf['All'] = dfdf['All'].astype(int)
dfdf['Pct'] = dfdf['M'] / dfdf['All'] * 100
dfdf.sort_values(by='Pct',ascending=False)
dfdf = dfdf.reset_index(drop=False)
dfdf
```

### 레버리지 큰 값 제거

```python
# OLS
model = sm.OLS.from_formula("units ~ date + store_nbr + item_nbr + day + year + "
                                 "month + weekday + is_weekend + is_holiday + is_holiday_weekday + "
                                 "is_holiday_weekend + holiday_name + around_BlackFriday + station_nbr + "
                                 "tmax + tmin + tavg + depart + dewpoint + wetbulb + heat + cool + sunrise + "
                                 "sunset + snowfall + preciptotal + stnpressure + sealevel + resultspeed + resultdir"
                                 "+ avgspeed", data=df)
result = model.fit()

# 레버리지 평균값
influence = result.get_influence()
hat = influence.hat_matrix_diag
print("레버리지의 평균 :", hat.mean())

# 레버리지 평균값의 4배를 넘지 않는 데이터만 남긴다.
idx = hat < (hat.mean() * 4)
df_result = df.iloc[idx, :]
df_result
```

### 아웃라이어 제거

```python
# OLS
model_2 = sm.OLS.from_formula("units ~ date + store_nbr + item_nbr + day + year + "
                                 "month + weekday + is_weekend + is_holiday + is_holiday_weekday + "
                                 "is_holiday_weekend + holiday_name + around_BlackFriday + station_nbr + "
                                 "tmax + tmin + tavg + depart + dewpoint + wetbulb + heat + cool + sunrise + "
                                 "sunset + snowfall + preciptotal + stnpressure + sealevel + resultspeed + resultdir"
                                 "+ avgspeed", data=df_result)
result_2 = model_2.fit()

# 표준화 잔차의 평균값
mean_sr = np.mean(result_2.resid_pearson)

# 표준화 잔차의 평균값의 4배보다 큰 값 제거(작은 값 남기기)
idx = result_2.resid_pearson < (mean_sr * 4)
df_result2 = df_result.iloc[idx, :]
df_result2
```
